## power transformation 
- log transformation: change of scales. Log distance closer than the linear distance.
- good to use when you need to change the scale of the graph (see the difference above between the old hist from 'avggift' to the new one.

```python
def log_transform_clean(x):
    x=np.log(x)
    if np.isfinite(x):
        return x
    else:
        return 0
df['avg_gift_log'] = list(map(log_transform_clean,df['avggift']))
```
