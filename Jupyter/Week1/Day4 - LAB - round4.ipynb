## Dropping the vehicle_type column, because has 50,25% of empty values.

```python
df = data.drop(columns='vehicle_type')
data = df
```

```python
data.reset_index(drop=True, inplace=True)
```

## Changing categorical empty values to 'unknow'

```python
data['state']=data['state'].fillna("unknow")
data['vehicle_class']=data['vehicle_class'].fillna("unknow")
data['vehicle_size']=data['vehicle_size'].fillna("unknow")
data['response']=data['response'].fillna("unknow")
```

```python
nulls_data1=pd.DataFrame(round(data.isna().sum()/len(data),4)*100)
nulls_data1=nulls_data1.rename(columns={'index':'header',0:'prop_nulls'})
```

## Dealing columns with numeric numbers

```python
med_months=data['months_since_last_claim'].median()
med_complaints=data['number_of_open_complaints'].median()
```

```python
data['months_since_last_claim']=data['months_since_last_claim'].fillna(med_months)
data['number_of_open_complaints']=data['number_of_open_complaints'].fillna(med_complaints)
```

```python
nulls_data2=pd.DataFrame(round(data.isna().sum()/len(data),4)*100)
nulls_data2=nulls_data2.rename(columns={'index':'header',0:'prop_nulls'})
nulls_data2
```

```python
# Using Seaborn
n_rows=3
n_cols=3
# Using FOR to create the subplots
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(15, 15))
for i, column in enumerate(df_num.columns):
    sns.histplot(df_num[column],ax=axes[i//n_cols,i%n_cols])
```

## check the multicollinearity between the features.
- For the numerical variables, check the multicollinearity between the features. Please note that we will use the column total_claim_amount later as the target variable.

```python
corre_matrix=df_num.corr()
corre_matrix
```

```python
mask=np.zeros_like(corre_matrix)
mask[np.triu_indices_from(mask)]=True
fig, ax = plt.subplots(figsize=(10,8))
ax=sns.heatmap(corre_matrix, mask=mask, annot=True, cmap=sns.diverging_palette(20, 220, n=2000));
```
